## 1 Remove html tags 

```python
sample = '''<ol>
<li><i>Hundreds of Beavers (2022)</i></li>
<li><i>Supermarket Woman (1996)</i></li>
<li><i>River (2023)</i></li>
<li><i>Tokyo Olympiad (1965)</i></li>
<li><i>Slacker (1990)</i></li>
<li><i>Dance With Me (2019)</i></li>
<li><i>The Big Clock (1948)</i></li>
<li><i>A Taxing Woman (1987)</i></li>
<li><i>Something Wild (1986)</i></li>
<li><i>The Fastest Gun Alive</i> (1956)
</li></ol></p></span> </p>'''
```

```python
sample
```

```python
import re
```

```python
# function that remove html tag
def removeHTML(data):
    p = re.compile(r'<.*?>')
    # p = re.sub(r"<.*?>", "", data)
    return p.sub('', data)
```

```python
sample = removeHTML(sample)
sample
```
