```python
import  pandas as pd
df = pd.DataFrame({'text' :['people like harshil', 'harshil like people', 'people make happy', 'harshil make happy'], 'output': [1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
bow = cv.fit_transform(df.text)
```

```python
cv.get_feature_names_out()
```

```python
bow.toarray()
```

```python
cv.vocabulary_
```

```python
cv.transform(["this is harshil people happy"]).toarray()
```

```python
cv = CountVectorizer(ngram_range = (1, 2))
bow = cv.fit_transform(df.text)
```

```python
cv.vocabulary_
```

```python
bow.toarray()
```

```python
cv.transform(["this is harshil people happy"]).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid = TfidfVectorizer()
tfid.fit_transform(df.text).toarray()
```

```python
tfid.idf_
```

```python
tfid.get_feature_names_out()
```
